```python
import time
import random
from datetime import datetime

# Configurações
LIMITE_CRITICO = 80  # Porcentagem de uso
INTERVALO_CHECK = 2  # Segundos entre verificações
ALERTA_ATIVO = False  # Para evitar disparos repetidos sem mudança de estado

def enviar_alerta(valor):
    """Simula o disparo de um e-mail ou alerta de sistema."""
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print(f"\n[ALERTA CRÍTICO] - {timestamp}")
    print(f"MOTIVO: Uso de CPU atingiu {valor}%")
    print("AÇÃO: Notificação enviada ao administrador via E-mail Simulado.\n")

def monitorar_sistema():
    global ALERTA_ATIVO

    print(f"--- Iniciando Monitoramento (Limite: {LIMITE_CRITICO}%) ---")

    try:
        while True:
            # Simulando a captura de uma métrica real (ex: psutil.cpu_percent())
            uso_atual = random.randint(30, 100)

            # Lógica de Disparo Único
            if uso_atual > LIMITE_CRITICO:
                if not ALERTA_ATIVO:
                    enviar_alerta(uso_atual)
                    ALERTA_ATIVO = True
            else:
                if ALERTA_ATIVO:
                    print(f"--- Sistema Normalizado ({uso_atual}%) ---")
                ALERTA_ATIVO = False

            time.sleep(INTERVALO_CHECK)

    except KeyboardInterrupt:
        print("\nMonitoramento encerrado pelo usuário.")

if __name__ == "__main__":
    monitorar_sistema()
```

```python
import time
import logging
import random
from datetime import datetime

# --- CONFIGURAÇÕES DO SISTEMA ---
CONFIG = {
    "THRESHOLD": 85.0,
    "CHECK_INTERVAL": 3,
    "SAMPLES_REQUIRED": 3,  # O erro deve persistir por 3 ciclos para alertar
}

class AlertaEngine:
    """Responsável unicamente pela comunicação de saída."""
    @staticmethod
    def disparar(valor, metrica="CPU"):
        timestamp = datetime.now().strftime("%H:%M:%S")
        print(f"\n🚀 [SISTEMA DE ALERTA CRÍTICO]")
        print(f"HORA: {timestamp} | MÉTRICA: {metrica} | VALOR: {valor}%")
        print("MENSAGEM: Condição crítica persistente detectada. Verifique o servidor.")
        print("-" * 40 + "\n")

class MonitorHardware:
    """Responsável pela lógica de monitoramento e decisão."""
    def __init__(self, config):
        self.config = config
        self.contador_critico = 0
        self.alerta_ativo = False

        logging.basicConfig(level=logging.INFO, format='%(levelname)s: %(message)s')

    def capturar_metrica(self):
        """Simula a leitura de um sensor (Modularizado para fácil troca)."""
        return round(random.uniform(70, 100), 2)

    def processar_logica(self, valor):
        """Aplica a lógica condicional e debouncing."""
        if valor >= self.config["THRESHOLD"]:
            self.contador_critico += 1
            logging.warning(f"Anomalia detectada: {valor}% ({self.contador_critico}/{self.config['SAMPLES_REQUIRED']})")
        else:
            if self.alerta_ativo:
                logging.info(f"Sistema estabilizou em {valor}%")
            self.contador_critico = 0
            self.alerta_ativo = False

    def verificar_disparo(self, valor):
        """Decide se deve disparar o alerta único."""
        if self.contador_critico >= self.config["SAMPLES_REQUIRED"] and not self.alerta_ativo:
            AlertaEngine.disparar(valor)
            self.alerta_ativo = True

    def iniciar(self):
        """Loop principal autônomo."""
        logging.info("Agente autônomo em execução...")
        try:
            while True:
                leitura = self.capturar_metrica()
                self.processar_logica(leitura)
                self.verificar_disparo(leitura)
                time.sleep(self.config["CHECK_INTERVAL"])
        except KeyboardInterrupt:
            print("\nEncerrando monitor...")

# --- PONTO DE ENTRADA ---
if __name__ == "__main__":
    monitor = MonitorHardware(CONFIG)
    monitor.iniciar()
```
